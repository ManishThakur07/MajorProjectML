IMPORTING LIBRARIES

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

READ CSV FILE

```python
df = pd.read_csv("creditcard.csv")
df
```

CHECKING NULL VALUES

```python
df.isnull().sum()
```

CHECKING INFO OF DATASET

```python
df.info()
```

VISUALISATION OF DATASET

```python
sns.heatmap(df)
```

DROP COLUMNS

```python
df = df.drop(columns=["Time","Class"])
```

SPLIT INPUT OUTPUT COLUMNS

```python
x = df.iloc[:,:28]
y = df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=34)
```

USE OF LINEARREGRESSION ALGORITHM

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predict = lr.predict(x_test)
```

```python
r2_score(y_test,predict)
```

USE OF DECISSIONTREEREGRESSOR ALGORITHM

```python
dtr = DecisionTreeRegressor()
```

```python
dtr.fit(x_train,y_train)
```

```python
dtrpredict = dtr.predict(x_test)
```

```python
r2_score(y_test,dtrpredict)
```

USE OF RANDOMFOREST REGRESSOR ALGORITHM

```python
rdr = RandomForestRegressor()
```

```python
rdr.fit(x_train,y_train)
```

```python
rdrpredict = rdr.predict(x_test)
```

```python
r2_score(y_test,rdrpredict)
```
